```python
import pandas as pd
hemo = pd.read_csv('/content/drive/MyDrive/Thesis Project Kidney/Hemodialysis.csv', encoding='latin1')
lab = pd.read_csv('/content/drive/MyDrive/Thesis Project Kidney/Lab.csv')
```

```python
patient_uid_list = hemo['PatientUID'].unique().tolist()
```

```python
# Remove patient UIDs from lab['PatientUID'] that are not in the list
lab = lab[~lab['PatientUID'].isin(patient_uid_list)]
```

## **WBC**

```python
selected_rows = lab[lab['ResultItemName'].notna() & lab['ResultItemName'].str.startswith('WBC')]
selected_rows
```

```python
unique_values = selected_rows['ResultItemName'].unique() #5
mapping = {value: 'WBC' for value in unique_values}

# Replace the unique values with 'WBC'
selected_rows['ResultItemName'] = selected_rows['ResultItemName'].replace(mapping)
```

```python
lab.update(selected_rows)
```
