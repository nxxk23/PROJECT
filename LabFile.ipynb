```python
import pandas as pd

# Function to clean and group by keywords
def clean_and_group(item):
    if item.startswith("CD"):
        return "CD4 count"
    elif item.startswith("Stool"):
        return "Stool"
    elif item == "Md Personnel Checkup stool examination":
        return "Stool"
    else:
        return item

# Modify the 'RequestItemName' column directly
lab['RequestItemName'] = lab['RequestItemName'].apply(clean_and_group)
lab['RequestItemName'] = lab['RequestItemName'].str.extract(r'^(.*?)( \(|$)')[0].str.strip()
```

```python
import pandas as pd

# Function to clean and group by keywords
def clean_and_group(item):
    if item.startswith("BODY FLUID"):
        return "BODY FLUID"
    elif item.startswith("Fluid:"):
        return "BILIRUBIN FLUID"
    else:
        return item

# Modify the 'RequestItemName' column directly
lab['RequestItemName'] = lab['RequestItemName'].apply(clean_and_group)
lab['RequestItemName'] = lab['RequestItemName'].str.extract(r'^(.*?)( \(|$)')[0].str.strip()
```

```python
import pandas as pd


# Function to clean and group by keywords
def clean_and_group(item):
    if pd.notna(item) and isinstance(item, str) and item.startswith("LH"):
        return "LH"
    else:
        return item

# Modify the 'RequestItemName' column directly
lab['RequestItemName'] = lab['RequestItemName'].apply(clean_and_group)
lab['RequestItemName'] = lab['RequestItemName'].str.extract(r'^(.*?)( \(|$)')[0].str.strip()

# Modify the 'ResultItemName' column directly
lab['ResultItemName'] = lab['ResultItemName'].apply(clean_and_group)
lab['ResultItemName'] = lab['ResultItemName'].str.extract(r'^(.*?)( \(|$)')[0].str.strip()
```
