- **Case 1 ขึ้นต้นด้วย ตัวเลข**

```python
drugCol = df['DrugName']
drugCol1 = [i.lower() for i in drugCol] #lower list
drugCol2 = [] #store drugname
for value in drugCol1:
    if value[0].isdigit():
        drugCol2.append(value)
    else:
        valList = value.split()
        #print(valList)
        # หาตำแหน่งของคำใน valList ที่ขึ้นต้นด้วยตัวเลข
        #ถ้ามีคำใดใน valList ที่ขึ้นต้นด้วยตัวเลขให้แสดงผลเป็นตำแหน่งคำนั้น
        #แต่ถ้าไม่มีให้แสดงผลเป็นความยาวของ valList เพื่อจะนำมาต่อเป็นคำเดิม เนื่องจากแยกคำแล้วไม่เจอตัวเลข
        index_begin_with_digit = len(valList)
        for i in range(len(valList)):
            if valList[i][0].isdigit():
                index_begin_with_digit = i
                break
        newval = " ".join(valList[:index_begin_with_digit])
        drugCol2.append(newval)
```

```python
start_with_bracket = df[df['DrugName'].str.startswith('[')]
unique_values = start_with_bracket['DrugName'].unique()
print(unique_values)
```

```python
# replace คำที่ไม่ต้องการด้วย
words_to_remove = ['mg','ml','tab','[7402]','[5702]','[5703]','[0604]','[5606]','[024840]']

drugCol3 = [i for i in drugCol2]
for word in words_to_remove:
    drugCol3 = [i.replace(word, "") for i in drugCol3]
```

```python
set(drugCol3)
```

```python
import pandas as pd

# Store values in drugCol2 into a DataFrame
df_drug = pd.DataFrame({'DrugName': drugCol2})

# Count unique values
value_counts = df_drug['DrugName'].value_counts()

# Save as CSV
value_counts.to_csv('unique_drug_counts.csv', index_label='DrugName')
```
