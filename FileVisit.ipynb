````python
import pandas as pd
hemo = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/Hemodialysis.csv', encoding='latin1')
````

````python
visit = pd.read_csv('/content/drive/MyDrive/ckdresearch2023/Visit.csv')
visit.head()
````

````python
# Create a list from hemo['PatientUID']
patient_uid_list = hemo['PatientUID'].unique().tolist()
# Remove patient UIDs from visit['PatientUID'] that are not in the list
visit = visit[~visit['PatientUID'].isin(patient_uid_list)]
visit.shape
````

````python
selected = visit[['PatientUID', 'PatientvisitUID', 'Sex', 'Age', 'ENTY', 'NATNL', 'Location', 'PricipalProblemCode', 'PricipalProblemName']]
selected
````

**Slope**

````python
import pandas as pd
slope = pd.read_excel('/content/drive/MyDrive/ckdresearch2023/Assign/SummaryFile/Slope.xlsx')


data = pd.DataFrame(slope)
data = data[['PatientUID','PatientVisitUID','Slopes']]
data = data.dropna()


df = pd.DataFrame(data)

# Sort the DataFrame by PatientUID and PatientVisitUID
df = df.sort_values(by=['PatientUID', 'PatientVisitUID'])

# Create a new column to store the previous slope value
df['PreviousSlope'] = df['Slopes'].shift(1)

# Update the Slopes column with the PreviousSlope values
df['Slopes'] = df['PreviousSlope']

# Drop the PreviousSlope column
df = df.drop(columns=['PreviousSlope'])
data = df.dropna()


data
````

````python
import pandas as pd

# Assuming you have a DataFrame named 'df' with your data
# You can create a new column 'InPatient' based on the 'ENTY' column
df = pd.DataFrame(visit)
df['InPatient'] = df['ENTY'].apply(lambda x: 1 if x == 'InPatient' else 0)
df['OutPatient'] = df['ENTY'].apply(lambda x: 1 if x == 'Out Patient' else 0)
df['Emergency'] = df['ENTY'].apply(lambda x: 1 if x == 'E' else 0)
df
````

````python
# Rename 'PatientUID' column in 'df' to 'PatientUID' to match 'table'
df.rename(columns={'PatientvisitUID': 'PatientVisitUID'}, inplace=True)

# Merge the two DataFrames based on 'PatientUID'
table = pd.merge(data, df, on=['PatientVisitUID','PatientUID'])
table
````
