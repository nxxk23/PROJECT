````python
import pandas as pd
prescription = pd.read_csv('/content/drive/MyDrive/Thesis Project Kidney/Prescription.csv')
prescription.head()
````

- **lower case and fill missing with space**

````python
prescription = prescription[['DrugName', 'GenericName']]
prescription['DrugName'] = prescription['DrugName'].str.lower()
prescription['GenericName'] = prescription['GenericName'].str.lower()
prescription['DrugName'].fillna(' ', inplace=True)
prescription['GenericName'].fillna(' ', inplace=True)
````

- **Delete [drugCode]**

````python
import pandas as pd

# Trim the part enclosed in square brackets at the front of 'DrugName' column
prescription['DrugName'] = prescription['DrugName'].str.replace(r'^\[[^\]]+\]\s*', '', regex=True)

# Trim the part enclosed in square brackets at the front of 'GenericName' column
prescription['GenericName'] = prescription['GenericName'].str.replace(r'^\[[^\]]+\]\s*', '', regex=True)
````

- **.tolist(**)

````python
drug = prescription['DrugName'].tolist()
````

````python
gen = prescription['GenericName'].tolist()
````

- **Remove Unit and junk words**

````python
drug_split = [[word for word in name.split() if word not in words_to_remove] for name in drug]
````

````python
gen_split = [[word for word in name.split() if word not in words_to_remove] for name in gen]
````

- **Remove digits** in list but kept the digits that in words

```
    ["chalktab-1.5", "1500", "ml"] => ["chalktab-1.5"]
    ["3tc", "lamivior", "100", "inj"] ==> ["3tc", "lamivior"]

```



````python
import re
drug_split = [[word for word in name if not re.match(r'^\d+(\.\d+)?|\d+/\d+$', word)] for name in drug_split]
````

````python
import re
gen_split = [[word for word in name if not re.match(r'^\d+(\.\d+)?|\d+/\d+$', word)] for name in gen_split]
````

- **Join into dataframe** with the same index

````python
import pandas as pd

drug_clean = [' '.join(name) for name in drug_split]

Drug = pd.DataFrame(drug_clean, columns=['DrugClean'])
````

````python
import pandas as pd

gen_clean = [' '.join(name) for name in gen_split]

Generic = pd.DataFrame(gen_clean, columns=['GenericClean'])
````

````python
df1 = prescription.merge(Drug[['DrugClean']], left_index=True, right_index=True)
df1
````

````python
df_final = df1.merge(Generic[['GenericClean']], left_index=True, right_index=True)
df_final
````

````python
df = df_final[['DrugClean','GenericClean']]
df
````

- **Count** the Unique Value

````python
pre_drug = df['DrugClean'].value_counts().reset_index()
pre_drug.columns = ['DrugName', 'Count']

pre_generic = df['GenericClean'].value_counts().reset_index()
pre_generic.columns = ['GenericName', 'Count']

pre_count = df.groupby(['DrugClean', 'GenericClean']).size().reset_index()
pre_count.columns = ['DrugName', 'GenericName', 'Count']
````

- Stored into **Excel**

````python
pre_df = pd.DataFrame(pre_count)
pre_drug = pd.DataFrame(pre_drug)
pre_generic = pd.DataFrame(pre_generic)
````

````python
with pd.ExcelWriter('PrescriptionCount.xlsx') as writer:
    pre_df.to_excel(writer, sheet_name='Prescription')
    pre_drug.to_excel(writer, sheet_name='CleanDrugName')
    pre_generic.to_excel(writer, sheet_name='CleanGenericName')
````
