```python
import pandas as pd
prescription = pd.read_csv('/content/drive/MyDrive/Thesis Project Kidney/Prescription.csv')
prescription.head()
```

```python
prescription = prescription[['DrugName', 'GenericName']]
prescription['DrugName'] = prescription['DrugName'].str.lower()
prescription['GenericName'] = prescription['GenericName'].str.lower()
prescription['DrugName'].fillna(' ', inplace=True)
prescription['GenericName'].fillna(' ', inplace=True)
prescription
```

```python
pre_list = prescription[['DrugName', 'GenericName']].values.tolist()
pre_list
```

```python
split_list = [[word for sublist in sublists for word in sublist.split(' ')] for sublists in pre_list]
```

```python
split_list
```

```python
drug_names = []
generic_names = []

for sublist in split_list:
    drug_name = ''
    generic_name = ''
    for word in sublist:
        if word.isnumeric():
            continue
        elif any(char.isdigit() for char in word):
            generic_name += word + ' '
        else:
            drug_name += word + ' '
    drug_names.append(drug_name.strip())
    generic_names.append(generic_name.strip())

df = pd.DataFrame({'DrugName': drug_names, 'GenericName': generic_names})
print(df)
```
